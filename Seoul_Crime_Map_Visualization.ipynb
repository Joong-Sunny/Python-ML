```python
import pandas as pd

github_base = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/master'
state_data = pd.read_csv(github_base+'/dataset/US_Unemployment_Oct2012.csv')
state_data.head()
```

```python
import json

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data = github_base+'/dataset/us-states.json',
    data=state_data, columns=['State', 'Unemployment'],
    nan_fill_color='purple', nan_fill_opacity=0.4,
    key_on='feature.id', fill_color='YlGn',
    legend_name='Unemployment Rate (%)'
).add_to(m)

m
```

```python
crime_path = '/content/drive/My Drive/Colab Notebooks/'+\
                    '데이터 사이언스 입문/Data/seoul_crime_result.csv'
geo_path = '/content/drive/My Drive/Colab Notebooks/데이터 사이언스 입문/'+\
            'skorea_municipalities_geo_simple.json'

crime_gu = pd.read_csv(crime_path, encoding='utf-8')
geo_str = json.load(open(geo_path, encoding='utf-8'))

crime_gu.head()
```

```python
crime_gu.sort_values(by="살인", ascending=False).head()
```

```python
my_map = folium.Map(location=[37.5502, 126.982], zoom_start=11, 
                    tiles='Stamen Toner')

folium.Choropleth(
    geo_data=geo_str,
    data=crime_gu, columns=['구', '살인'],
    nan_fill_color='purple', nan_fill_opacity=0.4,
    key_on='feature.id', fill_color='PuRd',
    legend_name='살인 발생 건수'
).add_to(my_map)

my_map
```
